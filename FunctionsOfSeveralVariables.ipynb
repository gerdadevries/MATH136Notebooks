# Functions of Several Variables

By now, we are very familiar with the calculus of single-variable functions such as $y = f(x)$ such as $y = 1 - x$, $y = x^2$, and $y = \sin(x)$.

In this notebook, we begin the study of multi-variable functions.  Mathematically, we are interested in a function of two variables, for example $z = f(x, y) = \sin(x+y)$, or in a function of three variables, for example $f(x, y, z) = x^2 + y^2 + z^2$, or more generally in a function of $n$ variables, $f( x_1, x_2, \ldots, x_n )$.  

### 1. Visualizing Functions of Two Variables: Surfaces and Contour Plots

Formally, a function of two variables, $f$, is a rule that assigns to each point $(x,y)$ in the domain $D$ a real number $z = f(x,y)$. Functions of two variables define <i>surfaces</i> in $\mathbb{R}^3$ which we often can visualize.

> If $f$ is a function of two independent variables with domain $D$, then the graph of $f$ is the set of all points $(x,y,z)$ such that $z = f(x,y)$ for all $(x,y)$ in $D$.  

Note that we now work with coordinate triplets, $(x,y,z)$.  We locate each point $(x,y,z)$ in $\mathbb{R}^3$ using the Cartesian coordinate system. An example of a surface visualized in $\mathbb{R}^3$ is the blue shape in the left figure below:

A complementary method to visualize functions of two variables is the contour plot.  A contour plot is like a topographical map showing the level curves (see the right figure above).
> A level curve of $z=f(x,y)$ consists of the points $(x,y)$ in the $xy$ plane with equations $f(x,y)=k$, where $k$ is a constant in the range of $f$.

In the left figure above, the level curves of the surface $z=f(x,y)$ are projected onto the $xy$ plane.  

In what follows, we present two specific examples.  For each example, we show the surface in $\mathbb{R}^3$ on the left, and level curves in the corresponding contour plot immediately to the right.

### 2. Visualization of a Function of Two Variables (Example 1)

Consider $$z= f(x,y) = x^2 + y^2.$$

The surface and the corresponding contour plot are shown below. 



<font size=0>Credit: Generated by WolframAlpha using the command `plot x^2 + y^2, x=-2..2 y=-2..2`.</font>

To connect the equation of the surface, $z=f(x,y)=x^2+y^2$, to its visualizations, recall that $x^2 + y^2 = R^2$ is the equation of a circle with radius $R$ centered at the origin in the $xy$ plane.  Therefore, each level curve given by $x^2 + y^2 = k$ is a circle with radius $\sqrt{k}$ centered at the origin in the $xy$ plane. Equivalently, the intersection of the horizontal plane $z=k$ (any plane that is parallel to the $xy$ plane) and the surface is a circle with radius $\sqrt{k}$.  The larger the value of $k$, the larger the circle.

The circles are clearly shown in the contour plot.  Here, seven level curves $x^2 + y^2 = k$ are shown, for $k=1, 2, \ldots, 7$.  The level curve given by $k=4$ is a circle of radius $2$ in the $xy$ plane.
> Take a moment to locate the level curve for $k=4$ in the contour plot, and verify that it is a circle passing through the points $(\pm 2,0)$ and $(0, \pm 2)$. 

The surface is called a <i>paraboloid</i>.  Note that the intersection of the surface with the $yz$ plane (defined by $x=0$) is the parabola $z = y^2$.  Similarly, the intersection of the surface with the $xz$ plane (defined by $y=0$) is the parabola $z=x^2$. 
> Take a moment to imagine the location of the $yz$ and $xz$ planes in the plot in $\mathbb{R}^3$ above, and verify that their intersections with the surface appear to be a parabolas.

### 5. Visualizing Functions of Three or More Variables

Unfortunately, we are not able to visualize functions of three or more variables very easily, as we would need to visualize in $\mathbb{R}^4$ or higher-dimensional space.  To gain insight into how these functions behave, one might think about the behaviour on certain convenient cross-sections by setting one or more variables equal to a constant (similar to what we did in the previous section, for example by considering the intersection of the horizontal plane $z=k$ with the surface).  This is not easy though, and certainly not a learning outcome for this course.

### 6. Summary

- If $f$ is a function of two independent variables with domain $D$, then the graph of $f$ is the set of all points $(x,y,z)$ such that $z = f(x,y)$ for all $(x,y)$ in $D$.


- A level curve of $z=f(x,y)$ consists of the points $(x,y)$ in the $xy$ plane with equations $f(x,y)=k$, where $k$ is a constant in the range of $f$.


- A contour plot is a graph in the $xy$ plane with a set of level curves of $z=f(x,y)$.

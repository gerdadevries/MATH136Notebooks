# Phase Plane Analysis

In this notebook, we develop tools to perform phase plane analysis, which is to obtain a qualitative understanding of the nature of the solutions of a system of two autonomous differential equations.  Conceptually, phase plane analysis is an extension of a phase plot analysis for a single autonomous differential equation into the two-dimensional phase plane.

### 1. Equilibria

Consider the following system of two autonomous differential equations:
$$\begin{eqnarray}
\frac{dx}{dt} & = f(x,y) \\
\frac{dy}{dt} & = g(x,y).
\end{eqnarray}$$

An <i>equilibrium</i> is a pair of values $(\hat{x},\hat{y})$ such that both $\displaystyle \frac{dx}{dt} = 0$ and $\displaystyle \frac{dy}{dt} = 0$ when $x=\hat{x}$ and $y = \hat{y}$, that is, when $f(\hat{x},\hat{y}) = 0$ and $g(\hat{x},\hat{y}) = 0$ simultaneously.

<b>Example:</b>

Find the equilibria of the system
$$\begin{eqnarray}
\frac{dx}{dt} & = x(3-x-y) \\
\frac{dy}{dt} & = y(2-x-y)
\end{eqnarray}$$
for $x,y \ge 0$.

$\displaystyle \frac{dx}{dt} = 0$ when $x(3-x-y) = 0$, that is, when 
1. $x=0$, or 
2. $y=3-x$.

Similarly, $\displaystyle \frac{dy}{dt} = 0$ when $y(2-x-y) = 0$, that is, when 
3. $y=0$, or 
4. $y=2-x$.

Combining conditions 1 and 3:
> This gives the equilibrium $(\hat{x},\hat{y}) = (0,0)$.

Combining conditions 1 and 4: 
> Substituting $x=0$ into $y=2-x$ gives $y=2$, resulting in the equilibrium $(\hat{x},\hat{y}) = (0,2)$.

Combining conditions 2 and 3:
> Substituting $y=0$ into $y=3-x$ gives $x=3$, resulting in the equilibrium $(\hat{x},\hat{y}) = (3,0)$.

Combining conditions 2 and 4:
> $y$ cannot simultaneously be equal to $3-x$ and $2-x$.  Therefore there is no equilibrium point resulting from combining conditions 2 and 4.

In summary, the given system of differential equations has three equilibria, namely $(0,0)$, $(0,2)$, and $(3,0)$.

### 2. Nullclines

Referring to the previous example, the curves $x=0$ and $y=3-x$ are the curves in the $(x,y)$ plane where $\displaystyle \frac{dx}{dt} = 0$.  We call these curves the $x$-<i>nullclines</i> of the given system of differential equations.  Similarly, the curves $y=0$ and $y=2-x$ are the curves where $\displaystyle \frac{dy}{dt} = 0$, and they are called the $y$-<i>nullclines</i>.  The following figure shows the nullclines.  For ease of recognition throughout, $x$-nullclines will be drawn in blue, and $y$-nullclines will be drawn in red.



In general, the $x$-<i>nullclines</i> of the system of differential equations 
$$\begin{eqnarray}
\frac{dx}{dt} & = f(x,y) \\
\frac{dy}{dt} & = g(x,y)
\end{eqnarray}$$
are the curves in the $(x,y)$ phase plane that satisfy the equation $f(x,y)=0$.  Along these curves, $\displaystyle \frac{dx}{dt} = 0$.  Similarly, the $y$-<i>nullclines</i> of the system of differential equations are the curves in the $(x,y)$ phase plane that satisfy the equation $g(x,y)=0$.  Along these curves, $\displaystyle \frac{dy}{dt} = 0$.

Any point at which an $x$-nullcline intersects a $y$-nullcline is an equilibrium.  In the following figure, we show the superposition of the two sets of nullclines in the phase plane, and indicate the equilibria with black dots.

### 4. Qualitative Dynamics in the Phase Plane

The $x$-nullclines separate regions where either $\displaystyle \frac{dx}{dt}>0$ (that is, where $x$ is increasing) or $\displaystyle \frac{dx}{dt}<0$ (that is, where $x$ is decreasing).

Similarly, $y$-nullclines separate regions where either $\displaystyle \frac{dy}{dt}>0$ (that is, where $y$ is increasing) or $\displaystyle \frac{dy}{dt}<0$ (that is, where $y$ is decreasing).

In the following figure, we use horizontal arrows to indicate the regions where $x$ is increasing (right-pointing arrows) or decreasing (left-pointing arrows), and vertical arrows to indicate the regions where $y$ is increasing (upward-pointing arrows) or decreasing (downward-pointing arrows). 

 

> NOTE: To determine the direction of a horizontal arrow in a particular region, choose a test point inside that region, and determine whether $\displaystyle \frac{dx}{dt} = f(x,y) < 0$ or $>0$.  For example, in the figure on the left, the point $(1,1)$ lies below the $x$-nullcline $y=3-x$.  We evaluate $f(1,1) = 1 (3-1-1) > 0$, therefore $x$ is increasing in the region below the $x$-nullcline $y=3-x$, and the horizontal arrow below this nullcline is pointing to the right.  

> Similarly, to determine the direction of a vertical arrow in a particular region, choose a test point inside that region, and determine whether $\displaystyle \frac{dy}{dt} = g(x,y) < 0$ (resulting in an upward-pointing arrow) or $>0$ (resulting in a downward-pointing arrow). 

Superimposing the nullclines and arrows results in the following figure, where the black arrows give the overall direction of movement of a solution trajectory.  
 


> NOTE: On any $x$-nullcline, $x$ does not change, and the movement is purely vertical.  Similarly, on any $y$-nullcline, $y$ does not change, and the movement is purely horizontal.

Following the general direction of the arrows, we can infer solution trajectories from several initial conditions, as shown in the figure below (the solution trajectories are sketched in green).



> NOTE: It appears that most initial conditions result in a solution that eventually ends up at $(3,0)$.  We therefore expect the equilibrium $(3,0)$ to be locally stable, and both $(0,0)$ and $(0,2)$ to be unstable.  

> Any initial condition $(0,y_0)$ for $y_0 > 0$ results in a solution that eventually ends up at $(0,2)$.  However, for any small deviation to the right off the $y$-axis, the solution ends up at $(3,0)$ instead.

### 5. An Online Tool for Qualitative Dynamics in the Phase Plane

Ariel Barton at the University of Arkansas has created a basic but functional tool to draw many directional arrows in the phase plane for a given set of differential equations in the form 
$$\begin{eqnarray}
\frac{dx}{dt} & = f(x,y) \\
\frac{dy}{dt} & = g(x,y).
\end{eqnarray}$$

The tool can be found at https://aeb019.hosted.uark.edu/pplane.html

It is possible to choose the number of direction arrows.  For example, setting "Number of arrows" to 20 results in 20 arrows across as well as 20 arrows down (for a total of 400 direction arrows).  The more arrows, the easier it is to infer the flow of the solutions.

### 6. Caution

While we were able to infer the stability of the equilibria for the system of differential equaitions in our example, it is often not possible from the direction arrows.

Particularly trickly are situations where there appears to be circular motion around a equilibrium, such as shown in the following figure.

 

> Note that there may be a periodic solution around the equilibrium (as shown in the middle), or the solution could spiral in towards the equilibrium (as shown on the left), or out from the equilibrium (as shown on the right).  The equilibrium in the figure on the left is locally stable, while the equilibrium in the figure on the right is unstable.  The equilibrium in the middle figure is neither (it is a degenerate case, referred to as neutrally stable).  

Determining the stability of equilibria in general is <b>beyond the scope of MATH 136</b> (this requires techniques from Linear Algebra, and is treated in a course of ordinary differential equations [MATH 334] and courses on mathematical modelling [MATH 371 and 372]).

### 7. Summary

We consider the system of two autonomous differential equations
$$\begin{eqnarray}
\frac{dx}{dt} & = f(x,y) \\
\frac{dy}{dt} & = g(x,y).
\end{eqnarray}$$

- An <i>equilibrium</i> of the system is a pair of values $(\hat{x},\hat{y})$ such that both $\displaystyle \frac{dx}{dt} = 0$ and $\displaystyle \frac{dy}{dt} = 0$ when $x=\hat{x}$ and $y = \hat{y}$, that is, when $f(\hat{x},\hat{y}) = 0$ and $g(\hat{x},\hat{y}) = 0$ simultaneously.


- The $x$-<i>nullclines</i> of the system are the curves in the $(x,y)$ phase plane that satisfy the equation $f(x,y)=0$.  Along these curves, $\displaystyle \frac{dx}{dt} = 0$.  Similarly, the $y$-<i>nullclines</i> are the curves in the $(x,y)$ phase plane that satisfy the equation $g(x,y)=0$.  Along these curves, $\displaystyle \frac{dy}{dt} = 0$.


- Any intersection of an $x$-nullcline and a $y$-nullcline is an equilibrium.


- The $x$-nullclines separate regions where either $\displaystyle \frac{dx}{dt}>0$ (that is, where $x$ is increasing) or $\displaystyle \frac{dx}{dt}<0$ (that is, where $x$ is decreasing).  Similarly, $y$-nullclines separate regions where either $\displaystyle \frac{dy}{dt}>0$ (that is, where $y$ is increasing) or $\displaystyle \frac{dy}{dt}<0$ (that is, where $y$ is decreasing).
    - Putting this information together in each region in the phase plane allows for a qualitative understanding of the nature of the solutions in the phase plane.

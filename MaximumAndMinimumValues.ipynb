# Maximum and Minimum Values

In this notebook, we use partial derivatives to find extreme values  of functions of two variables.  

### 1. Definitions: Local vs Absolute Maxima and Minima

Recall that a function of two variables, $z=f(x,y)$, defines a surface. An example of such surface, with local and absolute extrema indicated, is shown below.



<font size=0>Credit: J. Stewart and T. Day, Biocalculus: Calculus for the Life Sciences, Cengage Learning, 2015.</font>

We define local and absolute extrema as follows:

- A function $f(x,y)$ has a <b>local maximum</b> at $(a,b)$ if $f(x,y) \le  f(a,b)$ when $(x,y)$ is near $(a,b)$.  The number $(a,b)$ is called a local maximum value.

- A function $f(x,y)$ has a <b>local minimum</b> at $(a,b)$ if $f(x,y) \ge  f(a,b)$ when $(x,y)$ is near $(a,b)$.  The number $(a,b)$ is called a local minimum value.

- If $f(x,y) \le f(a,b)$ for all $(x,y)$ in the domain of $f$, then $f$ has an <b>absolute maximum</b> value at $(a,b)$.

- If $f(x,y) \ge f(a,b)$ for all $(x,y)$ in the domain of $f$, then $f$ has an <b>absolute minimum</b> value at $(a,b)$.

### 2. Characteristics of Local Extrema

Recall that a function of one variable, a local extremum may be found at $c$ if $f'(c) = 0$ (if there is a horizontal tangent line at $c$) or if $f'(c)$ does not exist (if there is no tangent line, for example, at the vertex of the graph of $y=|x|$).

This idea carries over nicely to functions of two variables.  Instead of tangent lines, we must consider tangent planes.  The orientation of a tangent plane at $(a,b)$ is determined by the partial derivatives $f_x(a,b)$ and $f_y(a,b)$.  The tangent plane is horizontal (parallel to the $xy$ plane) if both $f_x(a,b)= 0$ and $f_y(a,b)=0$. 

For a function of two variables then, a local extremum may be found at $(a,b)$ if both $f_x(a,b)= 0$ and $f_y(a,b)=0$ (if there is a horizontal tangent plane at $(a,b)$) or if $f_x(a,b)$ or $f_y(a,b)$  does not exist (if there is no tangent plane at $(a,b)$).

Examples of three surfaces with a local extremum at $(0,0)$ are shown below.

For the first two examples, the surfaces are smooth at $(0,0)$, that is, both $f_x(0,0)$ and $f_y(0,0)$ exist.  Moreover, these partial derivatives are 0, and there is a horizontal tangent plane at $(0,0)$.  For the first example, there is a local minimum at $(0,0)$; for the second example, there is a local maximum at $(0,0)$.

For the third example, there is a local maximum at $(0,0)$.  However, the surface is not smooth there; it has a sharp vertex at $(0,0)$ and there is no tangent plane at $(0,0)$.  Neither $f_x(0,0)$ nor $f_y(0,0)$ exists.

### 3. Critical Points

Recall that for a function of one variable, $y=f(x)$, a number $c$ in the domain of $f(x)$ is a critical number of $f$ if $f'(c)=0$ or $f'(c)$ does not exist.

Similarly, for a function of two variables, $z=f(x,y)$: 
>A point $(a,b)$ in the domain of $f(x,y)$ is a <b>critical point</b> of $f(x,y)$ if $f_x(a,b)=0$ and $f_y(a,b)=0$, or if one (or both) of these partial derivatives does not exist.

Recall that for a function of one variable, we have to be careful not to make the conclusion that there is a local extremum at $c$ if $f'(c)=0$.  Instead, there may be a point of inflection.  For example, think of $c=0$ for $f(x) = x^3$.

Similarly for functions of two variables: we cannot make the conclusion that there is a local extremum at $(a,b)$ if both $f_x(a,b)=0$ and $f_y(a,b)=0$.  Instead, there may be a <i>saddle point</i>, such as illustrated in the following figure, showing the surface $f(x,y) = x^2 - y^2$.



<font size=0>Credit: Generated by WolframAlpha using the command `plot x^2-y^2`.</font>

Note that $f_x = 2x = 0$ when $x=0$ and $f_y = -2y = 0$ when $y=0$.  That is, $(0,0)$ is a critical point.  However, there is no local extremum at $(0,0)$, but a saddle point (where the surface increases in the direction of $x$ but decreases in the direction of $y$).

### 3. Characterizing Critical Points

The Second Derivative Test for functions of two variables is analogous to the Second Derivative Test for functions of one variable.  We state it here without proof or justification (this is beyond the scope of this course):

>Suppose that the second partial derivatives of $f(x,y)$ are continous on a disk centered at the point $(a,b)$, where $f_x(a,b) = 0$ and $f_y(a,b) = 0$.  Let $$D(x,y) = f_{xx} f_{yy} - f_{xy}^2.$$
1. If $D(a,b) > 0$ and $f_{xx}(a,b) < 0$, then $f(a,b)$ is a local maximum.
2. If $D(a,b) > 0$ and $f_{xx}(a,b) > 0$, then $f(a,b)$ is a local minimum.
3. If $D(a,b) < 0$, then $f$ has a saddle point at $(a,b)$.
4. If $D(a,b) = 0$, then the test is inconclusive: $f$ could have a local maximum or minimum at $(a,b)$, or $(a,b)$ could be a saddle point of $f$.

### 4. Finding Absolute Absolute Maximum and Minimum Values

For functions of one variable, the existence of absolute extrema is guaranteed on a closed interval.  To find the absolute extrema, we can use the Closed Interval Method, which instructs us to evaluate the function at all critical points and at the endpoints of the interval, and then select the largest and smallest functions values.

The Closed Interval Method naturally extends to functions of two variables.  Instead of a working on a closed interval, we need to work on a closed, bounded set (a closed set in $\mathbb{R}^2$ is a set that includes its boundary; a bounded set is one that can be contained within a disk of finite radius).  An example of a closed, bounded set is the rectangle defined by all $(x,y)$ such that $0 \le x \le 3$ and $0 \le y \le 2$.
 
>To find the absolute maximum and minimum values of a continuous function $f(x,y)$ on a closed, bounded set $D$ in $\mathbb{R}^2$:
1. Determine the values of $f$ at all critical points of $f$ in $D$.
2. Find the extreme values of $f$ on the boundary of $D$.
3. The largest of the values from steps 1 and 2 is the absolute maximum; the smallest of these values is the absolute minimum.

### 5. Summary

<b>Local maximum and minimum values:</b>

- A function $f(x,y)$ has a local maximum at $(a,b)$ if $f(x,y) \le  f(a,b)$ when $(x,y)$ is near $(a,b)$.  The number $(a,b)$ is called a local maximum value.


- A function $f(x,y)$ has a local minimum at $(a,b)$ if $f(x,y) \ge  f(a,b)$ when $(x,y)$ is near $(a,b)$.  The number $(a,b)$ is called a local minimum value.

<b>Absolute maximum and minimum values:</b>

- If $f(x,y) \le f(a,b)$ for all $(x,y)$ in the domain of $f$, then $f$ has an absolute maximum value at $(a,b)$.


- If $f(x,y) \ge f(a,b)$ for all $(x,y)$ in the domain of $f$, then $f$ has an absolute minimum value at $(a,b)$.

<b>Critical points:</b>

- An interior point $(a,b)$ in the domain of $f(x,y)$ is called a critical point of $f(x,y)$ if $f_x(a,b)=0$ and $f_y(a,b)=0$, or if one (or both) of these partial derivatives does not exist.

    - If $f(x,y)$ has a local maximum or minimum at $(a,b)$, then $(a,b)$ is a critical point of $f$.
    
    - Critical points are candidates for local maximum and minimum values; if $(a,b)$ is a critical point of $f(x,y)$, there may or may not be a local maximum or local minimum at $(a,b)$.
    
<b>Second Derivative Test:</b>

- Suppose that the second partial derivatives of $f(x,y)$ are continous on a disk centered at the point $(a,b)$, where $f_x(a,b) = 0$ and $f_y(a,b) = 0$.  Let $$D(x,y) = f_{xx} f_{yy} - f_{xy}^2.$$
    1. If $D(a,b) > 0$ and $f_{xx}(a,b) < 0$, then $f(a,b)$ is a local maximum.
    2. If $D(a,b) > 0$ and $f_{xx}(a,b) > 0$, then $f(a,b)$ is a local minimum.
    3. If $D(a,b) < 0$, then $f$ has a saddle point at $(a,b)$.
    4. If $D(a,b) = 0$, then the test is inconclusive: $f$ could have a local maximum or minimum at $(a,b)$, or $(a,b)$ could be a saddle point of $f$.

<b>Finding absolute maximum and minimum values:</b>

- To find the absolute maximum and minimum values of a continuous function $f(x,y)$ on a closed, bounded set $D$ in $\mathbb{R}^2$:
    1. Determine the values of $f$ at all critical points of $f$ in $D$.
    2. Find the extreme values of $f$ on the boundary of $D$.
    3. The largest of the values from steps 1 and 2 is the absolute maximum; the smallest of these values is the absolute minimum.

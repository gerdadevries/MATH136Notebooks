### 1. Previous Encounters with Limits

We have come across limits previously in two different contexts, namely in the definition of the derivative, and in determining the behaviour sequences and functions at infinity.

#### (a) Limits in the Definition of the Derivative 

Recall that the derivative of a function $f$ at $x$ was defined as $$f'(x) = \lim_{h \to 0} \frac{f(x+h)-f(x)}{h}$$ provided that this limit exists.  Without knowing the rules of differentiation, determining the derivative $f'(x)$ for a given $f(x)$ is not obvious on first glance, since both the numerator and denominator approach $0$ and $\displaystyle \frac{0}{0}$ is not defined.  When we encountered this type of limit in the past, we were able to evaluate it after simplifying the difference quotient through algebraic manipulation.  For example, for $\displaystyle f(x) = x^2$, we have 
\begin{align}
f'(x) 
& = \lim_{h \to 0} \frac{ (x+h)^2 - x^2 }{h} \\
& = \lim_{h \to 0} \frac{ (x^2 + 2 x h + h^2) - x^2 }{ h } \\
& = \lim_{h \to 0} \frac{ 2xh + h^2 }{h} \\ 
& = \lim_{h \to 0} \frac{ h (2x + h ) }{h} \\ 
& = \lim_{h \to 0} \ (2x + h) \\
& = 2x.
\end{align}
The first four limits on the right-hand side are all of the type $\displaystyle \frac{0}{0}$.  By algebraic manipulation, we see that we can divide both the numerator and denominator by $h$, after which we can finally evaluate the limit, and see that for $f(x) = x^2$, we have $f'(x) = 2x$.

#### (b) Limits at Infinity to Determine the Behaviour of a Function at Infinity

In determining the shape or graph of a function $f(x)$, it often is helpful to know how the function behaves for large values of $x$.  For example, consider the absorption function $$\alpha(c) = \frac{4c}{1+c},$$ where $c$ is the concentration.
> An absorption function $\alpha(c)$ describes the total amount of chemical absorbed as a function of the concentration $c$.   

Both the numerator and denominator increase to infinity as $c \to \infty$, and so here we have a limit of the type $\displaystyle \frac{\infty}{\infty}$.  With algebraic manipulation (namely dividing numerator and denominator by the highest power of $c$ that occurs in the denominator), we can evaluate this limit as follows:
\begin{align}
\lim_{c \to \infty} \frac{4c}{1+c} 
& = \lim_{c \to \infty} \frac{\frac{4c}{c}}{\frac{1}{c}+\frac{c}{c}}  \\ 
& = \lim_{c \to \infty} \frac{ 4 }{ \frac{1}{c} + 1 } \\
& = \frac{ 4 }{ \lim_{c \to \infty} \frac{1}{c} + 1 } \\
& = \frac{ 4 }{ 0 + 1 } \\ 
& = 4.
\end{align}
We can conclude that the function $\alpha(c)$ has a horizontal asymptote at 4, that is, absorption saturates at 4, as shown in the following graph:

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
c = np.linspace(0, 50, 1000)
plt.plot(c, 4*c/(1+c), label=r'$\alpha(c) = \frac{4c}{1+c}$')
plt.xlim(0,50)
plt.ylim(0,4)
plt.xlabel('Concentration $c$', fontsize=15)
plt.ylabel(r'Absorption $\alpha(c)$', fontsize=15)
plt.legend(loc='best', fontsize=15)
plt.show()
```

### 2. Some Problematic Limits

There are many instances where algebraic manipulations are of no help to determine a limit of the type $\displaystyle \frac{0}{0}$ or of the type $\displaystyle \frac{\infty}{\infty}$.  We refer to these limits as <b>indeterminate form of type</b> $\displaystyle \frac{0}{0}$ and <b>indeterminate form of type</b> $\displaystyle \frac{\infty}{\infty}$, respectively.  This is especially the case when trigonometric, exponential, and/or logarithmic functions are involved.

For example, $$\lim_{x \to 0} \frac{e^x - 1}{x}$$ has <b>indeterminate form of type</b> $\displaystyle \frac{0}{0}$ because both the numerator and denominator approach 0 as $x \to 0$.  

Similarly, $$\lim_{x \to \infty} \frac{ \ln(x) }{ x }$$ has <b>indeterminate form of type</b> $\displaystyle \frac{\infty}{\infty}$ because both the numerator and denominator go to infinity as $x \to \infty$.

Neither limit can be evaluated with the methods that we have studied to date.

### 4. More Problematic Limits

In addition to limits that have indeterminate forms of the type $\displaystyle \frac{0}{0}$ or $\displaystyle \frac{\infty}{\infty}$, there are five other indeterminate forms, namely
- type $0 \cdot \infty$ (indeterminate product);
- type $\infty - \infty$ (indeterminate difference);
- types $0^0$, $\infty^0$, and $1^{\infty}$ (indeterminate powers).

With algebraic manipulations, each of these forms can be rewritten as an indeterminate form of the type $\displaystyle \frac{0}{0}$ or $\displaystyle \frac{\infty}{\infty}$, after which l'Hospital's Rule can be applied.  In the following video, we work through a few examples of this.

### 5. Summary

#### L'Hospital's Rule

>Suppose $f$ and $g$ are differentiable functions, and $g'(x) \ne 0$ near $a$ (except possibly at $a$).  Further, suppose that <br>
$$ \lim_{x \to a} f(x) = 0 \ \ \ \mbox{and} \ \ \ \lim_{x \to a} g(x) = 0 $$
or 
$$ \lim_{x \to a} f(x) = \pm \infty \ \ \ \mbox{and} \ \ \ \lim_{x \to a} g(x) = \pm \infty. $$
With other words, we have an indeterminate form of type $\displaystyle \frac{0}{0}$ or $\displaystyle \frac{\infty}{\infty}$. 
<br>
Then 
$$ \lim_{x \to a} \frac{ f(x) }{ g(x) } = \lim_{x \to a} \frac{ f'(x) }{ g'(x) }.$$

You should be able to recognize seven different types of indeterminate forms:
- Indeterminate Fractions: $\displaystyle \frac{0}{0}$ and $\displaystyle \frac{\infty}{\infty}$
- Indeterminate Product: $0 \cdot \infty$
- Indeterminate Difference: $\infty - \infty$
- Indeterminate Powers: $0^0$, $\infty^0$, and $1^{\infty}$

Remember that:
- L'Hospital's Rule directly apply ONLY to indeterminate <b>fractions</b> ($\displaystyle \frac{0}{0}$ and $\displaystyle \frac{\infty}{\infty}$). 
- In the case of an indeterminate <b>product</b> ($0 \cdot \infty$) or indeterminate <b>difference</b> ($\infty - \infty$), first use algebraic manipulation to rewrite the indeterminate form as an indeterminate fraction, and then apply l'Hospital's Rule.
- In the case of an indeterminate <b>power</b> ($0^0$, $\infty^0$, or $1^{\infty}$), first rewrite the limit via $\displaystyle A = e^{\ln(A)}$, then interchange the exponential function and the limit so that you can apply the limit to $\ln(A)$, and then use the laws of logarithms to simplify $\ln(A)$ (resulting in the limit of an indeterminate product).

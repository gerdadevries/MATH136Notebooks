In this notebook, we embark on the study of probability of continuous random variables (such as the height of a randomly chosen MATH 136 student, or the cholesterol level of a randomly chosen female living in Edmonton).  It turns out that calculus plays an important role in working with continuous random variables. ### 1. A Very Brief Review of Probability

Recall that probability is the extent to which an event is likely to occur.  Said differently, it is the ratio of the number of favourable outcomes to all possible outcomes. 
> For example, flipping a fair coin has two possible outcomes: Heads or Tails.  The probability of obtaining heads is 1/2, and we write $P$(heads) = $\frac{1}{2}$.

> Similarly, rolling a fair die has six possible outcomes: 1, 2, 3, 4, 5, or 6.  The probability of rolling 4 is 1/6, and we write $P(4) = \frac{1}{6}$.

For any event $A$, we have $0 \le P(A) \le 1$.  
- If $P(A) = 0$, it is impossible for event $A$ to occur.
- If $P(A) = 1$, event $A$ will occur with certainty.  

If $A^c$ is the complement of $A$, then $P(A^c) = 1 - P(A)$.
> For example, for a fair die, the probability of not rolling 4 is $P($not $4) = P(1$ or $2$ or $3$ or $5$ or $6) = 1 - P(4) = \frac{5}{6}$.
 
### 2. Random Variables: Discrete versus Continuous

Outcomes of random experiments, such a rolling a die or measuring the cholesterol level of a randomly chosen individual, usually are real numbers.  We describe such outcomes by <b>random variables</b>.

A <i>random variable</i> $X$ is a variable whose value may change with a random experiment.  We write it as a capital letter to distinguish it from the usual variables that we have been using throughout the course.  

If a random variable can take on only a finite number of values, we say it is a <i>discrete</i> random variable.  
> If we let $X$ be the outcome of rolling a fair die, then the possible values of $X$ are 1, 2, 3, 4, 5, and 6, and $P(X=4) = \frac{1}{6}$.

If a random variable can take on any value (in an interval of real numbers), we say it is a <i>continuous</i> random variable.
> If we let $X$ be the height of a randomly chosen MATH 136 student, then the possible value of $X$ are anywhere between 50 and 250 cm.  Note that although we typically take a height measurement to the nearest cm, height itself can take on any value between 50 and 250 cm.  

From now on, we concern ourselves only with continuous random variables.

### 3. Probability Density Functions (PDFs)  

Any continuous random variable $X$ has a <i>probability density function</i> $f$ that satisfies the following three conditions:
1. $f(x) \ge 0$.
2. $\displaystyle \int_{-\infty}^{\infty} f(x) \ dx = 1.$
3. $\displaystyle P(a \le X \le b) = \int_a^b f(x) \ dx.$
    
Conditions 1 and 2 are needed because we are working with probabilities.  Condition 3 implies that the probability that $X$ takes on a value between $a$ and $b$ is the <i>area</i> under the probability density function between $a$ and $b$.

>For example, the probability that a randomly chosen MATH 136 student would be between 176 and 180 cm tall is $$\int_{176}^{180} f(x) \ dx,$$ where $f$ is the probability density function associated with the height of a randomly chosen MATH 136 student.  

For many continuous random variables, the probability density function is nonzero only on some interval of the real numbers.  For example, suppose that we are analyzing a database of soil samples where the concentration of nitrogen (in g/kg) is known not to exceed 0.5 g/kg (and cannot be smaller than 0 g/kg).  The function   
$$f(x) = \begin{cases} 192 x \left( \frac{1}{2} - x \right)^2 & \text{when $0 \le x \le 0.5$} \\
0 & \text{otherwise}\end{cases}$$
thus is a reasonable candidate to be the probability density function for the concentration of nitrogen (in g/kg) in a randomly selected soil sample from the database.

### 4. Cumulative Distribution Functions (CDFs)

The <i>cumulative distribution function</i> for the continuous random variable $X$ is the function $F(x)$ defined by 
$$F(x) = P( X \le x ).$$

Geometrically, $F(x)$ is the area under the graph of the probability density function to the left of the value $X = x$, namely
$$F(x) = \int_{-\infty}^x f(s) \ ds.$$

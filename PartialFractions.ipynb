## Partial Fractions

In this notebook, we introduce another essential tool in the evaluation of integrals, namely the method of partial fractions.  

### 1. Motivation

It is true that 
$$
\frac{1}{x^2-16} = \frac{1}{(x-4)(x+4)} = \frac{1}{8} \left[ \frac{1}{x-4} - \frac{1}{x+4} \right].
$$

You can verify this by working backwards, that is, by starting from the right-hand side and making a common denominator (you should go ahead and do this - be sure to carefully keep track of negative signs).

Although we cannot readily integrate the left-hand side, we <b>can</b> readily integrate the right-hand side.  We then get 
\begin{align}
\int \frac{dx}{x^2-16} 
& = \frac{1}{8} \left[ \int \frac{dx}{x-4} - \int \frac{dx}{x+4} \right] \\ 
& = \frac{1}{8} \left[ \ln | x-4 | - \ln | x+4| \right] + C \\
& = \frac{1}{8} \ln \left| \frac{x-4}{x+4} \right| + C.
\end{align}

The key to making progress in the previous example was to write a 'complicated' rational function (in this case $\frac{1}{x^2-16})$ as a sum of simpler functions.

> Recall: A rational function is of the form $\displaystyle \frac{P(x)}{Q(x)}$, where both $P(x)$ and $Q(x)$ are polynomials.

Partial fraction decomposition, or simply 'partial fractions', is a systematic method to write a rational function as a sum of simpler functions.

### 2. Overview of Partial Fraction Decomposition

Consider a rational function $\displaystyle \frac{P(x)}{Q(x)}$ such that the degree of $P(x)$ <b>is less than</b> the degree of $Q(x)$.

Then the steps to determine the partial fraction decomposition are as follows:

1. Factor $Q(x)$ as far as possible.
> In MATH 136, we will restrict ourselves to the case where $Q(x)$ is a product of <b>distinct</b> linear factors.
2. Write the partial fraction decomposition using the rule that any linear factor $ax+b$ in $Q(x)$ gives rise to a partial fraction of the form $$\frac{A}{ax+b},$$ where $A$ is an unknown to be determined in step 3.
> Example: $$\frac{4x^2+14x-2}{(x-1)(x+1)(x+3)} = \frac{A}{x-1} + \frac{B}{x+1} + \frac{C}{x+3}.$$
3. Solve for the unknowns ($A$, $B$, $C$, etc.).

Once you have solved for the unknowns, you can proceed with any integration if needed.

### 5. CAUTION! A Long Division Step May Be Needed Before Proceeding with a Partial Fraction Decomposition

Up to now, we have considered rational functions $\displaystyle \frac{P(x)}{Q(x)}$ where the degree of $P(x)$ <b>is less than</b> the degree of $Q(x)$.

This raises the question what to do when the degree of $P(x)$ is the same or larger than the degree of $Q(x)$.
> Example: The rational function $$\frac{4x^2+3x+2}{x^2+x-2}$$ has a polynomial with degree 3 in the numerator and a polynomial with degree 2 in the denominator.  Since the degree in the numerator is larger than the degree in the denominator, we cannot proceed immediately with the partial fraction decomposition.

In these cases, it is necessary to take the preliminary step of using long division to divide $Q(x)$ into $P(x)$ and obtain  
$$
\frac{P(x)}{Q(x)} = S(x) + \frac{R(x)}{Q(x)},
$$
where $S(x)$ is a polynomial and $\displaystyle \frac{R(x)}{Q(x)}$ is a rational function such that the degree of $R(x)$ is less than the degree of $Q(x)$.  The partial fraction decomposition then can proceed for the rational function on the right-hand side, $\displaystyle \frac{R(x)}{Q(x)}$.

### 6. Summary

Partial fraction decomposition is a systematic method to write a <i>proper</i> rational function as a sum of simpler rational functions. 
> A <i>proper</i> rational function is a rational function such that the degree of the polynomial in the numerator <b>is less than</b> the degree of the polynomial in the denominator.
For example, $\displaystyle \frac{x}{x^3-1}$ is a proper rational function, 
but neither $\displaystyle \frac{x^3+2x-5}{x^3-1}$ 
nor $\displaystyle \frac{x^4-3x}{x^3-1}$ are.

In general, the simpler rational functions are of the form 
$$
\frac{A}{(ax+b)^n} \ \ \ \textrm{or} \ \ \ \frac{Bx+C}{(ax^2 + bx + c)^n},
$$
where $a$, $b$, $c$, $A$, $B$, and $C$ are constants and $n$ is a positive integer, and the quadratic $ax^2+bx+c$ is irreducible (cannot be factored further since $b^2 - 4 a c < 0$).

If a <i>proper</i> rational function is not already in one of these simpler forms, then the steps to determine the partial fraction decomposition are as follows:

1. Factor $Q(x)$ as far as possible.
 - In general, $Q(x)$ will be the product of linear factors and/or irreducible factors.
 - In MATH 136, we work with the restriction that $Q(x)$ is the product only of <i>distinct</i> linear factors.

2. Under the restriction on $Q(x)$ mentioned in Step 1, write the partial fraction decomposition using the rule that each linear factor $ax+b$ in $Q(x)$ gives rise to a partial fraction of the form $$\frac{A}{ax+b},$$ where $A$ is an unknown to be determined in step 3.

3. Solve for the unknowns ($A$, $B$, $C$, etc.).

Note: If the rational function $\displaystyle \frac{P(x)}{Q(x)}$ is not <i>proper</i>, then it is necessary to use long division first and obtain  
$$
\frac{P(x)}{Q(x)} = S(x) + \frac{R(x)}{Q(x)},
$$
where $S(x)$ is a polynomial and $\displaystyle \frac{R(x)}{Q(x)}$ is a <i>proper</i> rational function.

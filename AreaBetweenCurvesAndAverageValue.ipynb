# Definite Integrals for Area Between Curves and for Average Value of a Function

In this notebook, we briefly review (from the first calculus course) the main idea behind setting up the definite integral for computing the area of a region between two curves.  We will re-use this main idea in several upcoming notebooks.  We also review setting up the definite integral to compute the average value of a function on an interval.    

### 1. Review of Area Between Curves

Consider the region $S$ shown in the figure below.  The region is bounded above by the graph of $y=f(x)$, below by the graph of $y = g(x)$, to the left by the vertical line $x=a$, and to the right by the vertical line $x=b$.



Calculating $A$, the area of region $S$, is very similar to calculating the area under a curve, as shown in the figure below:
- We divide the interval $[a,b]$ into $n$ subintervals of equal width $\displaystyle \Delta x = \frac{b-a}{n}$.
- On each subinterval, we build an approximating rectangle with base $\Delta x$.
- The height of the $i$th approximating rectangle is $f(x_i^*) - g(x_i^*)$. 
- The area of the $i$th approximating rectangle is height times width, that is, $\left[ f(x_i^*) - g(x_i^*) \right] \Delta x$.
- The total area of all $n$ approximating rectangles is $\displaystyle \sum_{i=1}^n \left[ f(x_i^*) - g(x_i^*) \right] \Delta x$.
- The total area $A$ then is the limiting value of the latter, namely $$A = \lim_{n \to \infty} \sum_{i=1}^n \left[ f(x_i^*) - g(x_i^*) \right] \Delta x = \int_a^b \left[ f(x) - g(x) \right] \ dx.$$



### 2. Shortcut: Focus on a Typical Slice

### 3. Average Value of a Function

Many quantities are continuous. Examples include temperature at the Edmonton International Airport (as a function of time) and blood glucose concentration in a diabetic patient (as a function of time).  How would we calculate their average values over a period of time?

We know how to calculate the average of $n$ numbers, $y_1$, $y_2$, $\ldots$, $y_n$.  We add them up, and divide by $n$:
$$y_{ave} = \frac{ y_1 + y_2 + \ldots + y_n }{ n }.$$
To generalize this to the concept of an average value of a function, slicing (or chopping) again is going to be the key.

### 4. The Mean Value Theorem for Integrals

The Mean Value Theorem for Integrals guarantees that this is indeed the case (under the usual condition that the function is continuous).

### 5. Summary

#### Area between curves

The area $A$ of the region bounded by the curves of continuous functions $y = f(x)$ and $y = g(x)$, and the vertical lines $x=a$ and $x=b$, where $f(x) \ge g(x)$  for all $x$ in $[a,b]$, is $$A = \int_a^b \left[ f(x) - g(x) \right] \ dx.$$

#### Average value of a function

The average value of a function $y=f(x)$ on the interval $[a,b]$ is $$f_{ave} = \frac{1}{b-a} \int_a^b f(x) \ dx.$$

#### The Mean Value Theorem for Integrals

If $f$ is continuous on the interval $[a,b]$, then there exists a number $c$ such that 
$$f(c) = f_{ave} = \frac{1}{b-a} \int_a^b f(x) \ dx,$$
that is,
$$\int_a^b f(x) \ dx = f(c) \dot (b-a).$$
